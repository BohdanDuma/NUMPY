```python
countries = [
    "United States", "China", "India"
]
countrie = np.random.choice(countries,size=100)
order_id = np.arange(101,201)
id = np.arange(1,101)
t1 = {
    'id' : id,
    'country': countrie,  
       }
t2 = {
         'id': id,
         'amount': np.random.uniform(100,1100,100)
       }
t1 = pd.DataFrame(t1)
t2 = pd.DataFrame(t2)
```

```python
merged = pd.merge(t1,t2, on='id')
filtered = merged.groupby('country').agg(
    avg_amount=('amount','mean'),
    orders_count=('amount','count')
)
filtered = filtered[filtered['orders_count']>10]
filtered
```

```python
date = (f'{np.random.randint(2004,2015)}-{np.random.randint(1,12)}-{np.random.randint(1,31)}')
date_series = (f'{np.random.randint(2004,2015)}-{np.random.randint(1,12)}-{np.random.randint(1,31)}' for i in range(100))
date_series = pd.Series(date_series)
date_series = pd.to_datetime(date_series, errors='coerce')
df = { 'id' : np.arange(1,101),
         'date': date_series,
       }

df = pd.DataFrame(df)

df['year_month'] = df['date'].dt.to_period('M')

grouped = df.groupby('year_month')['id'].nunique()

grouped
```
