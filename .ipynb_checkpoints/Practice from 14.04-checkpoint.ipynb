```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


# Створюємо дані
data = {
    'item': ['Latte', 'Cappuccino', 'Latte', 'Tea', 'Espresso', 'Cappuccino', 'Latte', 'Tea', 'Cake', 'Cake'],
    'category': ['Drink', 'Drink', 'Drink', 'Drink', 'Drink', 'Drink', 'Drink', 'Drink', 'Food', 'Food'],
    'price': [50, 45, 50, 30, 35, 45, 50, 30, 80, 80],
    'quantity': [5, 12, np.nan, 8, 15, np.nan, 7, 10, 4, 6],
    'date': pd.to_datetime(['2026-04-01', '2026-04-01', '2026-04-02', '2026-04-02', '2026-04-03', '2026-04-03', '2026-04-04', '2026-04-04', '2026-04-05', '2026-04-05'])
}

df = pd.DataFrame(data)

```

```python
df.head
```

```python
df.shape
```

```python

df.columns
mask = df['quantity'].notna() 
mask[mask == True].count()
median = np.nanmedian(df['quantity'])
df['quantity'] = df['quantity'].fillna(median)
```

```python
df['total_sales'] = df['quantity'] * df['price']
```

```python

item_sales = df.groupby('item')['total_sales'].sum().reset_index()
```

```python
item_sales
```

```python
plt.figure(figsize =(12,6))
```

```python
plt.bar(item_sales['item'],item_sales['item'])
plt.ylabel('Вид напою')
plt.grid(axis='y', linestyle='--',alpha=0.3)
```

```python
daily_ravenue = df.groupby('date')['total_sales'].sum().reset_index()
```

```python
plt.plot(daily_ravenue['date'], daily_ravenue['total_sales'], 
         marker='D', linestyle='-', color='teal', linewidth=3, markersize=8)
plt.xticks(rotation=30)
```

```python
date_rev_drink = df[df['category'] == 'Drink'].groupby('date')['quantity'].sum().reset_index() 
```

```python
plt.plot(date_rev_drink['date'],date_rev_drink['quantity'])
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Створюємо "сирі" дані
raw_data = {
    'order_id': [101, 102, 103, 104, 105, 106, 107, 108, 109, 110],
    'product': ['Laptop', 'Mouse', 'Laptop', 'Monitor', 'Mouse', 'Keyboard', 'Laptop', 'Monitor', 'Mouse', 'Keyboard'],
    'category': ['Tech', 'Accessory', 'Tech', 'Tech', 'Accessory', 'Accessory', 'Tech', 'Tech', 'Accessory', 'Accessory'],
    'price': [1200, 25, 1200, 300, 25, 50, 1200, 300, 25, 50],
    'quantity': [1, 4, np.nan, 2, 10, np.nan, 1, 1, 5, 2],
    'date': ['2026-04-01', '2026-04-01', '2026-04-02', '2026-04-02', '2026-04-03', '2026-04-03', '2026-04-04', '2026-04-04', '2026-04-05', '2026-04-05'],
    'store': ['Lviv', 'Kyiv', 'Lviv', 'Odesa', 'Kyiv', 'Lviv', 'Odesa', 'Kyiv', 'Lviv', 'Odesa']
}

df = pd.DataFrame(raw_data)
```

```python
 df
```

```python
df['quantity'].fillna(np.nanmedian(df['quantity']))
df['date'] = pd.to_datetime(df['date'])
```

```python
df['revenue'] = df['price']* df['quantity']
```

```python
tech_only = df[(df['category'] == 'Tech') & (df['revenue'] > 500)]
sum_store = df.groupby('store')['revenue'].sum().reset_index()
avg_by_date = df.groupby('date')['quantity'].mean().reset_index()
```

```python
plt.bar(sum_store['store'], sum_store['revenue'])
```

```python
plt.plot(avg_by_date['date'], avg_by_date['quantity'])
plt.xticks(rotation=30)
```
