```python
import pandas as pd

products = pd.DataFrame({
    'product_id': [1, 2, 3],
    'product_name': ['Coffee', 'Tea', 'Cake']
})

sales = pd.DataFrame({
    'product_id': [1, 2, 1, 3, 2],
    'quantity': [2, 1, 5, 2, 3],
    'price': [50, 30, 50, 80, 30]
})
stock = pd.DataFrame({
    'product_id': [1, 2, 3],
    'count': [10, 0, 5] # Ой, Tea (ID 2) закінчився!
})
```

```python
new_merg = pd.merge(products, sales, on='product_id', how='left')
new_merg['revenue'] = new_merg['price'] * new_merg['quantity']
triple_merge = pd.merge( new_merg,stock, on='product_id', how='left')
new_merg.groupby('product_name')['revenue'].sum()
triple_merge
```

```python
lost_profit = triple_merge[triple_merge['count'] == 0]['revenue'].sum()
print(f"lost {lost_profit}")
```
