```python
from sklearn.linear_model import LogisticRegression # Не плутати з лінійною!
import numpy as np
# 1. Створюємо дані (ознаки: час на сайті в хв, кількість переглянутих сторінок)
X_train = np.array([
    [1, 1], [2, 1], [10, 5], [12, 8], [1, 2], [15, 10]
])

# 2. Ціль (y): 0 — не купив, 1 — купив
y_train = np.array([0, 0, 1, 1, 0, 1])
reg = LogisticRegression()
reg.fit(X_train, y_train)
reg.get_params(deep=True)
x= np.array([[7, 4]])

reg.predict_proba(x)
# Завдання:
# 1. Навчи модель LogisticRegression на цих даних.
# 2. Зроби прогноз для клієнта, який провів на сайті 5 хвилин і подивився 3 сторінки.
# 3. Використай метод model.predict_proba(), щоб побачити ЙМОВІРНІСТЬ покупки.
```

```python
from sklearn.metrics import classification_report

y_pred = reg.predict(X_train)
print(classification_report(y_train,y_pred))
```
