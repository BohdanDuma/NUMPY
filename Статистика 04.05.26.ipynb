### VARIANCE

```python
import pandas as pd
import numpy as np
from scipy import stats

d = np.array([1,2,3])
p = np.array([1/3,1/3,1/3])
sample = np.random.choice(d,size=1000, p=p)

mean = sample.mean()
std = (sample - mean)**2
D = std.mean()
print(mean, D)
```

```python
new_sample = sample * 5
mean = new_sample.mean()
std = (new_sample - mean)**2
D = std.mean()
print(mean, D)
```

### Bayes (Теорема Баєса)

якщо в нас умовна ймовірність(P(A|B), P(A), P(B)) то P(B|A) = P(A|B)*P(B)/P(A)

#### Задача (Класика інтерв'ю)
Уяви систему моніторингу пайплайнів:Хвороба (Помилка в даних): Трапляється рідко, лише у 1% випадків ($P(A) = 0.01$).Тест (Система алертингу):Якщо помилка є, алерт спрацьовує у 99% випадків ($P(B|A) = 0.99$).Якщо помилки немає, алерт все одно може спрацювати (хибна тривога) у 1% випадків ($P(B|\text{not } A) = 0.01$).👉 Питання: Тобі прийшов алерт (подія $B$ сталася). Яка реальна ймовірність того, що в даних дійсно є помилка?

```python
p_b = 0.99 * 0.01 + 0.01 * 0.99

p_ab = (0.99 * 0.01)/p_b
```

```python
print(p_ab)
```

де: (P(H|A)\) — Апостеріорна ймовірність (що ми хочемо дізнатися).

(P(A)\) — Повна ймовірність події 

(A\): загальна ймовірність появи доказу (як при істинній, так і при хибній гіпотезі).
